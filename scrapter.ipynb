```python
import warnings
import pandas as pd
import requests
from urllib.request import Request, urlopen

warnings.simplefilter(action='ignore', category=FutureWarning) # Hide FutureWarnings

# URL for Liverpool
team_url = 'https://fbref.com/en/squads/822bd0ba/2025-2026/matchlogs/c9/shooting/Liverpool-Match-Logs-Premier-League'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36'}

# Sending HTTP Get-request to webpage
request = Request(team_url, headers=headers)
data = urlopen(request).read()

# # Check for success response
# if response.status_code == 200:
#     print("Successfully retrieved page!")
# else:
#     print(f"Failed to retrieve page. Status code: {response.status_code}" )

# print(data.content.decode())
# print(data)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(data, "html.parser") # create a beautifulSoup Object for dealing HTML
print(soup.prettify()[:500])
```

```python
# find all tables in HTML
tables = soup.find_all("table")


print(tables)

for table in tables:
    print(table.get("id")) # print all id
```
